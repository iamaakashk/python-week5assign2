```python
try :
    
    d = { 'key' : 'ask' , 1 : [2,3,4,6] }
    
    print (d['key2'])
except KeyError as e :
    print(e)
```

```python
try :
    
    l = [2,3,4,5]
    
    print (l[6])
    
except IndexError as e :
    
    print(e)
```
